```python
import requests

# Ваш токен API, который вы получили на сайте Bitly

def shorten_url(long_url):
    """
    Функция для сокращения длинной URL с использованием Bitly API.

    :param long_url: Длинная URL, которую нужно сократить.
    :return: Сокращенная URL или None в случае ошибки.
    """
    headers = {
        'Authorization': f'Bearer {BITLY_ACCESS_TOKEN}',
        'Content-Type': 'application/json'
    }
    data = {
        'long_url': long_url
    }
    response = requests.post('https://api-ssl.bitly.com/v4/shorten', json=data, headers=headers)
    if response.status_code == 200:
        return response.json().get('link')
    else:
        print(f"Ошибка при сокращении ссылки: {response.status_code}")
        print(response.json())
        return None

# Пример использования функции
long_url = "https://docs.google.com/document/d/1M46-35Rfn5MuAH3LKjt4scDFD7d507ra/edit?usp=drive_link&ouid=118252777260935189446&rtpof=true&sd=true"
short_url = shorten_url(long_url)
if short_url:
    print(f"Сокращенная ссылка: {short_url}")
```

```python
import fitz  # PyMuPDF
from docx import Document

def load_pdf(file_path):
    """
    Извлекает текст из PDF документа.

    :param file_path: Путь к PDF файлу
    :return: Содержимое PDF документа в виде строки
    """
    try:
        print(f"Чтение PDF файла: {file_path}")
        doc = fitz.open(file_path)
        text = []
        for page_num in range(doc.page_count):
            page = doc.load_page(page_num)
            text.append(page.get_text())
        return "\n".join(text)
    except Exception as e:
        raise ValueError(f"Ошибка при чтении PDF файла: {str(e)}")

def load_docx(file_path):
    """
    Извлекает текст из DOCX документа.

    :param file_path: Путь к DOCX файлу
    :return: Содержимое DOCX документа в виде строки
    """
    try:
        print(f"Чтение DOCX файла: {file_path}")
        doc = Document(file_path)
        text = []
        for paragraph in doc.paragraphs:
            text.append(paragraph.text)
        return "\n".join(text)
    except Exception as e:
        raise ValueError(f"Ошибка при чтении DOCX файла: {str(e)}")
```

Пример кода для загрузки PDF документа

```python
from PyPDF2 import PdfReader

def load_pdf(file_path):
    try:
        with open(file_path, "rb") as pdf_file:
            reader = PdfReader(pdf_file)
            text = []
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                text.append(page.extract_text())
            return "\n".join(text)
    except FileNotFoundError:
        return f"Файл не найден: {file_path}"
    except Exception as e:
        return f"Ошибка при чтении PDF файла: {str(e)}"

pdf_file_path = r"E:\Для GPT\Git\ConsultantIntegrator\ConsultantIntegrator\20_PMKS_t01_Lektsiya.pdf"
pdf_content = load_pdf(pdf_file_path)
print(pdf_content)
```

Преимущества и недостатки
Преимущества: Гибкость, возможность обработки каждого формата отдельно.
Недостатки: Более сложный код, необходимость обработки ошибок для каждого типа документа.
